```python
import pandas as pd 
df = pd.read_csv("logs/performance_log.csv")
```

```python
df
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10, 5))
x= df['Gameweek']
y1 = df['Accuracy']
y2 = df['Log_Loss']
y3 = df['Brier_H']
y4 = df['Brier_D']
y5 = df['Brier_A']
ax.plot(x, y1, label = 'Accuracy')
ax.plot(x, y2, label = 'Log_Loss')
ax.plot(x, y3, label = 'Brier_H')
ax.plot(x, y4, label = 'Brier_D')
ax.plot(x, y5, label = 'Brier_A')
ax.legend()
plt.show()
```
